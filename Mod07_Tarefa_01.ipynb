#### 2) Vamos montar um metadados

1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.
2. Adicione uma coluna nesse *dataframe* chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável.
    Dica:
        1. inicie uma lista vazia
        2. faça um for ao longo dos nomes das variáveis,
        3. conte o número de categorias dessa variável
        4. acumule essa informação de 3. na lista que você criou em 1.
        5. No final, essa lista pode ser atribuída à nossa variável.

````python
demo01_df = pd.DataFrame(demo01)

# Criar o DataFrame de metadados
metadados = pd.DataFrame({
    "variavel": demo01_df.columns,
    "tipo_dado": [demo01_df[col].dtype for col in demo01_df.columns]
})

# Adicionar a coluna qtd_categorias
qtd_categorias = []
for col in demo01_df.columns:
    qtd_categorias.append(demo01_df[col].nunique())  # Conta as categorias únicas

metadados["qtd_categorias"] = qtd_categorias

metadados_df = pd.DataFrame(metadados)
metadados_df
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
demo01_df = pd.get_dummies(demo01_df
                           , columns = ['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia', 'mau']
                           , drop_first = True
                          )
````

````python
demo01_df.info()
````

````python
# Verificando as colunas booleanas no DataFrame
bool_columns = demo01_df.select_dtypes(include=bool).columns

# Convertendo booleanos em dummies (0 ou 1)
demo01_df[bool_columns] = demo01_df[bool_columns].astype(int)

# Exibindo o DataFrame atualizado
demo01_df
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
tab_email = (pd.crosstab(
    demo01_df['possui_email'],
    demo01_df['mau_True'],
    normalize='index'
) * 100).round(2)

tab_veiculo = (pd.crosstab(
    demo01_df['posse_de_veiculo_Y'],
    demo01_df['mau_True'],
    normalize='index'
) * 100).round(2)

# Exibindo os resultados
print("Tabela cruzada - possui_email (%):")
print(tab_email)

print("\nTabela cruzada - posse_de_veiculo (%):")
print(tab_veiculo)

print('\nCom base nisso, considero que a variável POSSE_DE_VEICULO_Y é mais poderosa.')
````
